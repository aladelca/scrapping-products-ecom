```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from typing import Dict, Any, Optional
from selenium import webdriver
import time
```

```python

def get_data(webpage):
    data = {
    "description": [],
    "brand": [],
    "original_price": [],
    "offer_price": []
    }
    driver = webdriver.Chrome()
    driver.get(webpage)
    driver.maximize_window()
    time.sleep(5)
    
    for i in range(1, 52):
        # Scraping description
        try:
            description = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/h3")
            data["description"].append(description.text)
        except:
            data["description"].append(None)
        
        # Scraping brand
        try:
            brand = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/span[1]")
            if brand.get_attribute("class") == "poly-component__brand":
                data["brand"].append(brand.text)
            else:
                try:
                    brand = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/span")
                    if brand.get_attribute("class") == "poly-component__brand":
                        data["brand"].append(brand.text)
                    else:
                        try:
                            brand = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/span[2]")
                            if brand.get_attribute("class") == "poly-component__brand":
                                data["brand"].append(brand.text)
                            else:
                                data["brand"].append(None)
                        except:
                            data["brand"].append(None)
                except:
                    data["brand"].append(None)
        except:
            data["brand"].append(None)
        
        # Scraping original price
        try:
            original_price = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/div[2]/s")
            if "previous" in original_price.get_attribute("class"):
                data["original_price"].append(original_price.text)
            else:
                data["original_price"].append(None)
        except:
            data["original_price"].append(None)
        
        # Scraping offer price
        try:
            offer_price = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/div[2]/div/span[1]")
            if offer_price.get_attribute("class") == "andes-money-amount andes-money-amount--cents-superscript":
                data["offer_price"].append(offer_price.text)
            else:
                try:
                    offer_price = driver.find_element(By.XPATH, f"/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/div[3]/div/span[1]")
                    print("I am here")
                    data["offer_price"].append(offer_price.text)
                except:
                    data["offer_price"].append(None)
        except:
            data["offer_price"].append(None)
    return data
```

```python
list_data = []
for i in range(1, 11):
    data_page = get_data(f"https://listado.mercadolibre.com.pe/ropa-accesorios/calzado/zapatillas/{i}")
    list_data.append(data_page)
```
